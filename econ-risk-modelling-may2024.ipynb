```python
import numpy as np
import matplotlib.pyplot as plt


T = 1.0
dt = 1/252
N = int(T / dt)
t = np.linspace(0, T, N)

mu_stock_normal = 0.05
mu_stock_risky = 0.03
sigma_stock_normal = 0.12
sigma_stock_risky = 0.25

mu_yield_normal = 0.0435  
mu_yield_risky = 0.05 
theta_yield = 0.5  
sigma_yield_normal = 0.01  
sigma_yield_risky = 0.02  


S0 = 100  
Y0 = 0.0435  


np.random.seed(42)
dW_stock = np.random.randn(N) * np.sqrt(dt)
dW_yield = np.random.randn(N) * np.sqrt(dt)

# Simulation arrays
stock_prices_normal = np.zeros(N)
stock_prices_risky = np.zeros(N)
yields_normal = np.zeros(N)
yields_risky = np.zeros(N)

stock_prices_normal[0] = S0
stock_prices_risky[0] = S0
yields_normal[0] = Y0
yields_risky[0] = Y0

for i in range(1, N):
    stock_prices_normal[i] = stock_prices_normal[i-1] * (1 + mu_stock_normal * dt + sigma_stock_normal * dW_stock[i])
    stock_prices_risky[i] = stock_prices_risky[i-1] * (1 + mu_stock_risky * dt + sigma_stock_risky * dW_stock[i])
    yields_normal[i] = yields_normal[i-1] + theta_yield * (mu_yield_normal - yields_normal[i-1]) * dt + sigma_yield_normal * dW_yield[i]
    yields_risky[i] = yields_risky[i-1] + theta_yield * (mu_yield_risky - yields_risky[i-1]) * dt + sigma_yield_risky * dW_yield[i]

fig, ax = plt.subplots(2, 1, figsize=(12, 10))

ax[0].plot(t, stock_prices_normal, label='Normal Risk (Stock Index)', linewidth=2)
ax[0].plot(t, stock_prices_risky, label='Increased Risk (Stock Index)', linestyle='--', linewidth=2)
ax[0].set_title('Simulated Stock Market Index Under Different Risk Perceptions Using Brownian Motion')
ax[0].set_xlabel('Time (years)')
ax[0].set_ylabel('Stock Market Index Value')
ax[0].legend()

ax[1].plot(t, yields_normal, label='Normal Risk (Yield)', linewidth=2)
ax[1].plot(t, yields_risky, label='Increased Risk (Yield)', linestyle='--', linewidth=2)
ax[1].set_title('Simulated Treasury Yield Under Different Risk Perceptions Using Brownian Motion')
ax[1].set_xlabel('Time (years)')
ax[1].set_ylabel('Yield')
ax[1].legend()

plt.tight_layout()
plt.show()
```
